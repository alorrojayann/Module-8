```python
# 1. With the earthquakes.csv file, select all the earthquakes in Japan with a magType of mb and a magnitude of 4.9 or greater

japan_df = earthquake_df[(earthquake_df['parsed_place'] == 'Japan') & (earthquake_df['magType'] == 'mb') & (earthquake_df['mag'] >= 4.9)]
japan_df
```

```python
# 2. Create bins for each full number of magnitude (for example, the first bin is 0-1, the second is 1-2, and so on)
# with a magType of ml and count how many are in each bin.

bin_edge = list(range(int(earthquake_df['mag'].max()) + 1))

mag_bin = pd.cut(earthquake_df['mag'], bins=bin_edge, right=False)

mag_bin.value_counts().sort_index()
```

```python
# 3. Using the faang.csv file, group by the ticker and resample to monthly frequency. Make the following aggregations:
#    - Mean of the opening price
#    - Maximum of the high price
#    - Minimum of the low price
#    - Mean of the closing price
#    - Sum of the volume traded

faang_df = pd.read_csv('/content/data/faang.csv', index_col='date', parse_dates=True)
faang_df.head(10)
```

```python
faang_df.info()
```

```python
# group by the ticker and resample to monthly frequency
monthly_faang = faang_df.groupby('ticker').resample('M').last()
monthly_faang.head(10)
```

```python
# Mean of the opening price
# Maximum of the high price
# Minimum of the low price
# Mean of the closing price
# Sum of the volume traded
import numpy as np

monthly_faang.agg({
    'open': np.mean,
    'high': np.max,
    'low': np.min,
    'close': np.mean,
    'volume': np.sum
})
```

```python
# 4. Build a crosstab with the earthquake data between the tsunami column and the magType column. Rather than showing the frequency count,
# show the maximum magnitude that was observed for each combination. Put the magType along the columns.
earthquake_df.head(10)
```

```python
pd.crosstab(
    index=earthquake_df.tsunami,
    columns=earthquake_df.magType,
    values='mag', # specify column to use
    aggfunc='max' # takes the max mag for each combination
)
```

```python
# 5. Calculate the rolling 60-day aggregations of OHLC data by ticker for the FAANG data. Use the same aggregations as exercise no. 3.
import numpy as np

cols=['open', 'high', 'low', 'close', 'volume']

faang_df.groupby('ticker')[cols].rolling(window='60D').agg({
    'open': np.mean,
    'high': np.max,
    'low': np.min,
    'close': np.mean,
    'volume': np.sum
})
```

```python
# 6. Create a pivot table of the FAANG data that compares the stocks.
# Put the ticker in the rows and show the averages of the OHLC and volume traded data.
cols=['open', 'high', 'low', 'close', 'volume']

faang_df.pivot_table(index='ticker', values=cols+['volume'], aggfunc='mean')
```

```python
# 7. Calculate the Z-scores for each numeric column of Netflix's data (ticker is NFLX) using apply().
nflx_data = faang_df[faang_df['ticker'] == 'NFLX']

# convert columns to numeric
num_columns = ['open', 'high', 'low', 'close', 'volume']
nflx_data[num_columns] = nflx_data[num_columns].apply(pd.to_numeric)

nflx_data[num_columns].apply(lambda x: (x - x.mean()) / x.std())
```

```python
# 8. Add event descriptions:
#    - Create a dataframe with the following three columns: ticker, date, and event. The columns should have the following values:
#       - ticker: 'FB'
#       - date: ['2018-07-25', '2018-03-19', '2018-03-20']
#       - event: ['Disappointing user growth announced after close.', 'Cambridge Analytica story', 'FTC investigation']
#    - Set the index to ['date', 'ticker']
#    - Merge this data with the FAANG data using an outer join

data = {
    'ticker': ['FB'] * 3,
    'date': ['2018-07-25', '2018-03-19', '2018-03-20'],
    'event': ['Disappointing user growth announced after close.',
              'Cambridge Analytica story',
              'FTC investigation']
}

event_df = pd.DataFrame(data)
event_df
```

```python
faang_df_reset = faang_df.reset_index()

event_df.set_index(['date', 'ticker'], inplace=True)

merged_faang = pd.concat([faang_df_reset, event_df], axis=1, join='outer')
merged_faang_df = merged_faang.reset_index()
merged_faang_df
```

```python
# 9. Use the transform() method on the FAANG data to represent all the values in terms of the first date in the data.
# To do so, divide all the values for each ticker by the values for the first date in the data for that ticker.
# This is referred to as an index, and the data for the first date is the base (https://ec.europa.eu/eurostat/statistics-explained/index.php/
# Beginners:Statisticalconcept-Indexandbaseyear). When data is in this format, we can easily see growth over time.
# Hint: transform() can take a function name.

def first_date(group):
  return group / group.iloc[0]

first_date_faang = faang_df.groupby('ticker').transform(first_date)
first_date_faang.head(10)
```
